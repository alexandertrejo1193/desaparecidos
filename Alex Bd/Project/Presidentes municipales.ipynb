```python
from bs4 import BeautifulSoup 
import requests 
import pandas as pd
stranio=['2002-2004','2005-2007','2008-2010','2011-2013','2014-2016','2017-2019','2019-2021','2021-2024']
df=pd.DataFrame({})
for anio in stranio:
    url='https://es.wikipedia.org/wiki/Anexo:Presidentes_municipales_de_Aguascalientes_({})'
    response = requests.get(url.format(anio))
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find_all('table', attrs={'class': "wikitable"})[1]
    df1 = pd.read_html(str(table))[0]
    df1["Periodo"]=anio
    df1["Estado"]="AGUASCALIENTES"
    df=pd.concat([df, df1], sort=True,ignore_index=True) 
    
```
