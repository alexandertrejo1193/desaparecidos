```python
import os
import pandas as pd

carpeta = '/Users/alexandertrejo/Documents/Lotus Analytica/Desaparecidos/Alex/desaparecidos/Alex Bd/Project/PresidentesMunicipales'

archivos = os.listdir(carpeta)

archivos_xlsx = [archivo for archivo in archivos if archivo.endswith('.xlsx')]

dataframes = []


for archivo in archivos_xlsx:
    ruta_archivo = os.path.join(carpeta, archivo)
    df = pd.read_excel(ruta_archivo)
    dataframes.append(df)

df_combinado = pd.concat(dataframes)

df_combinado['Año de inicio'] = df_combinado['Periodo'].str.extract(r'de (\d{4})')
df_combinado['Año de fin'] = df_combinado['Periodo'].str.extract(r'a (\d{4})')
df_combinado['Año de inicio'] = df_combinado['Año de inicio'].fillna(df_combinado['Periodo'].str.extract(r'del \d{2}-\w{3}-(\d{4})', expand=False))
df_combinado['Año de fin'] = df_combinado['Año de fin'].fillna(df_combinado['Periodo'].str.extract(r'al \d{2}-\w{3}-(\d{4})', expand=False))
df_combinado['Año de inicio'] = df_combinado['Año de inicio'].fillna(df_combinado['Periodo'].str.extract(r'(\d{4})', expand=False))

df_combinado.to_excel('/Users/alexandertrejo/Documents/Lotus Analytica/Desaparecidos/Alex/desaparecidos/Alex Bd/Project/combinado.xlsx',index=False)
df_combinado
```
